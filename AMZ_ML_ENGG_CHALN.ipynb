```python
import pandas as pd
import numpy as np
import sklearn as sk
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('training_data.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
dependent_var=data['customer_category']
customer_active_seg, X1 =[],[]
customer_active_seg=encoder.fit_transform(data.iloc[:,9].fillna('Nan'))
X1=encoder.fit_transform(data.iloc[:,10].fillna('Nan'))
encoded_df=data.drop(['customer_id','customer_active_segment','X1','customer_category'],axis=1)
encoded_df=encoded_df.fillna(0.0)
encoded_df['customer_active_segment']=customer_active_seg
encoded_df['X1']=X1

encoded_df.head()
```

```python
encoded_df.describe()
```

```python
train_len=int(.8*(encoded_df.shape[0]))
train_x,train_y=encoded_df[:train_len],dependent_var[:train_len]
test_x,test_y=encoded_df[train_len:],dependent_var[train_len:]
```

```python
from sklearn.ensemble import RandomForestClassifier

rfclf=RandomForestClassifier(random_state=0)

rfclf.fit(train_x,train_y)

print(f'training accuracyis:{rfclf.score(train_x,train_y)}')

pred4=rfclf.predict(test_x)

print(f'test accuracyis:{sk.metrics.accuracy_score(test_y,pred4)}')
```

```python
sub_data=pd.read_csv('testing_data.csv')
sub_data2=sub_data.set_index('customer_id')
sub_data2[0:5]
```

```python
sub_customer_active_seg, sub_X1 =[],[]
sub_customer_active_seg=encoder.fit_transform(sub_data2.iloc[:,8].fillna('Nan'))
sub_X1=encoder.fit_transform(sub_data2.iloc[:,9].fillna('Nan'))
sub_encoded_df=sub_data2.drop(['customer_active_segment','X1'],axis=1)
sub_encoded_df=sub_encoded_df.fillna(0.0)
sub_encoded_df['customer_active_segment']=sub_customer_active_seg
sub_encoded_df['X1']=sub_X1

sub_encoded_df.head()
```

```python
sub_customer_category=rfclf.predict(sub_encoded_df)
```

```python
ans=pd.DataFrame({'customer_id':sub_encoded_df.index,
                  'customer_category':sub_customer_category})
ans.to_csv('solution.csv',index=False)
```
